#### Filling null values

```python
def fill_gender(df):
    
    # filling null values of female column
    condition = df['Female'].isna()
    df.loc[condition,'Female'] = df.loc[condition,'Population'] - df.loc[condition,'Male']
    
    # filling null values of population column
    condition = df['Population'].isna()
    df.loc[condition,'Population'] = df.loc[condition,'Female'].add(df.loc[condition,'Male'])
#     df['Population'] = df['Population'].fillna(df['Population'].median())
    
    # filling null values of male column
    condition = df['Male'].isna()
    df.loc[condition,'Male'] = df.loc[condition,'Population'] - df.loc[condition,'Female']
    
    # return the filled dataframe
    return df
    
    
df1 = fill_gender(df1)
```

```python
def fill_literacy(df):
    # filling null values of female column
    condition = df['Literate_Female'].isna()
    df.loc[condition,'Literate_Female'] = df.loc[condition,'Literate'] - df.loc[condition,'Literate_Male']
#     df['Literate_Female'] = df['Literate_Female'].fillna(df['Literate_Female'].median())
    
    
    # filling null values of male column
    condition = df['Literate_Male'].isna()
    df.loc[condition,'Literate_Male'] = df.loc[condition,'Literate'] - df.loc[condition,'Literate_Female']
#     df['Literate_Male'] = df['Literate_Male'].fillna(df['Literate_Male'].median())

    # filling null values of population column
    condition = df['Literate'].isna()
    df.loc[condition,'Literate'] = df.loc[condition,'Literate_Female'].add(df.loc[condition,'Literate_Male'])
#     df['Literate'] = df['Literate'].fillna(df['Literate'].median())

    
    # return the filled dataframe
    return df

df1 = fill_literacy(df1)
```

```python
def fill_population(df):
    
    # filling null values of Age_Not_Stated column
    condition = df['Age_Not_Stated'].isna()
    df.loc[condition,'Age_Not_Stated'] = df.loc[condition,'Literate'] - df.loc[condition,'Young_and_Adult'] - df.loc[condition,'Middle_Aged'] - df.loc[condition,'Senior_Citizen'] 
#     df['Age_Not_Stated'] = df['Age_Not_Stated'].fillna(df['Age_Not_Stated'].median())
    
    
    # filling null values Young_and_Adult column
    condition = df['Young_and_Adult'].isna()
    df.loc[condition,'Young_and_Adult'] = df.loc[condition,'Literate'] - df.loc[condition,'Age_Not_Stated'] - df.loc[condition,'Middle_Aged'] - df.loc[condition,'Senior_Citizen'] 
#     df['Young_and_Adult'] = df['Young_and_Adult'].fillna(df['Young_and_Adult'].median())
    
    # filling null values of Middle_Aged column
    condition = df['Middle_Aged'].isna()
    df.loc[condition,'Middle_Aged'] = df.loc[condition,'Literate'] - df.loc[condition,'Young_and_Adult'] - df.loc[condition,'Age_Not_Stated'] - df.loc[condition,'Senior_Citizen'] 
#     df['Middle_Aged'] = df['Middle_Aged'].fillna(df['Middle_Aged'].median())
    
    # filling null values of Senior_Citizen column
    condition = df['Senior_Citizen'].isna()
    df.loc[condition,'Senior_Citizen'] = df.loc[condition,'Literate'] - df.loc[condition,'Young_and_Adult'] - df.loc[condition,'Age_Not_Stated'] - df.loc[condition,'Middle_Aged'] 
#     df['Senior_Citizen'] = df['Senior_Citizen'].fillna(df['Senior_Citizen'].median())
    
    # return the filled dataframe
    return df

df1 = fill_population(df1)
```

```python
def fill_household(df):
    
    # filling null values of Urban_Households column
    condition = df['Urban_Households'].isna()
    df.loc[condition,'Urban_Households'] = df.loc[condition,'Households'] - df.loc[condition,'Households_Rural']
#     df['Urban_Households'] = df['Urban_Households'].fillna(df['Urban_Households'].median())

    
    # filling null values of Households column
    condition = df['Households'].isna()
    df.loc[condition,'Households'] = df.loc[condition,'Urban_Households'].add(df.loc[condition,'Households_Rural'])
#     df['Households'] = df['Households'].fillna(df['Households'].median())
    
    
    # filling null values of Households_Rural column
    condition = df['Households_Rural'].isna()
    df.loc[condition,'Households_Rural'] = df.loc[condition,'Households'] - df.loc[condition,'Urban_Households']
#     df['Households_Rural'] = df['Households_Rural'].fillna(df['Households_Rural'].median())

    # return the filled dataframe
    return df


df1 = fill_household(df1)
```

### Problem Statement 11: (Create a function to alter the data to create uniformity)

```python
def convert_toint(x):
    if ',' in str(x):
        return int(x.replace(",", ""))
    elif pd.isna(x):
        return x
    else:
        return int(x)

def formating_(df):
    
    df['State/UT'] = df['State/UT'].apply(change_case)
    df = rep(df)
    
    # convert datatype from object to integer
    for i in df.columns:
        if df[i].dtypes !='int64' and i!='State/UT':
            
            df[i] = df[i].apply(convert_toint)
            
    return df

hospital = formating_(hospital)
hospital
```

### Problem Statement 16: (Gap in number of beds)

```python
#Making new dataframe with all india population
l= [["All India", (df1.groupby(['State/UT'])['Population'].sum().sum()) ]]
all_india_population = pd.DataFrame(l, columns = ['State/UT','Population'],index=[37])

# concating all India state with the total India's population row wise
all_india_population = pd.concat([df1.groupby(['State/UT'])['Population'].sum().reset_index(), all_india_population], axis=0).sort_values(by='State/UT')
(all_india_population)


# merging the Number of Beds in india to all india population
all_india_govt = pd.merge(left=all_india_population, right=govt_hospital[['State/UT','Total_Beds']] , on='State/UT')
all_india_govt
```

```python
# find the number of required beds
def exp_bed(x):
    
    if x>=3:
        return 0.1
    else:
        return 3-x

# getting total number of govt hospitals for 1000 people
all_india_govt['Available_bed_per1000_ratio'] = all_india_govt['Total_Beds']/all_india_govt['Population']*1000
all_india_govt['Expected_bed_per1000_ratio'] = all_india_govt['Available_bed_per1000_ratio'].apply(exp_bed)

# sorting based on Ratio
all_india_govt = all_india_govt.sort_values(by='Expected_bed_per1000_ratio',ascending=False)
```
